### operaciones basicas

````python
#calculmamos el precio por millas
df['price_per_miles'] = df.price / df.odometer
df.head()
````

### Trabajar con la funci√≥n lambda

````python
df['price_2'] = df.price.apply(lambda x : x*2)
df.price_2.head(15)
````

````python
# vamos a realizar otra columna donde lo multipicaremos el cada registro lo multiplicaremos por 3
df['price_3']= df.price.apply(lambda x: x*3)
df.price_3.head(15)
````

````python
# podemos crear grupos o jerarquias
df['grupos'] = df.price.apply( lambda x: 'expensives' if x > 10000 else 'cheap' )
df.grupos.head(15)
````

````python
# realizar la funcion en base a dos condiciones
df['new_group'] = df.apply(lambda x: 'yes' if x ['price'] > 10000 and x['age'] < 7 else 'no', axis =1)
df.new_group.tail()
````

````python
#otra  forma es
df['new_2'] = df[['price', 'age']].apply(lambda x: 'yes' if x[0]>10000 and x[1] <7 else 'no', axis ='columns')
df.new_2.head()
````

### Crear variables dummies

````python
df_dummmies = pd.get_dummies(df[['grupos', 'new_2']])
df_dummmies.head()
````

````python
# get dummies eliminando la primera columna
datos_dummies = pd.get_dummies(df[['grupos', 'new_2']], drop_first=True)
datos_dummies.head()
````

````python
#se puede realizar con una columna o mas de una a la vez
data_dummies = pd.get_dummies(df[['grupos', 'new_2', 'new_group']])
data_dummies.head()
````

### Pivot tablet

````python
df.pivot_table(index='year', columns ='type', values='price', aggfunc='mean').sort_index()
````

````python
df.pivot_table(index = 'year', columns='manufacturer', values = 'price', aggfunc='mean')
````

````python
# se puede aplicar una fucion fuera de esto lo haria una serie no es la forma correcta
df.pivot_table(index = 'year', columns = 'type', values = 'price' ).mean()
````

### Agrupar  
groupby y pivot son muy utilizados para realizar an√°lisis

````python
df.groupby(['year', 'type']).price.mean().unstack()
````

````python
df.groupby(['year', 'type']).price.mean()
````

````python
df.groupby('type').mean()
````

````python
df.groupby('manufacturer').count().unstack()
````

````python
#agrupar manufactura de vehiculo por a√±o traemos el precio promedio de vehiculos, eliminamos la columna de indice
df.groupby(['manufacturer', 'year'], as_index=False).price.mean()
````

````python
# si queremos saber cual es el precio pormedio mayor de vehiculos segun su modelo
mayor_precio = df.groupby('manufacturer').price.mean()
mayor_precio.nlargest(n=10)
````

### Unir - Combinar Registros  

```python
pd.merge()
pd.append()
pd.concat()
```

````python
# es igual a join que tambien es utilizado en python pero merge y concat son mas utilizados
#creamos dos DataFrame de ejemplos con un filtro sencillo
df1 = df[['url', 'region']]
df2 = df[['url', 'price']]

#merge

df_merged = pd.merge(df1, df2, on='url')
df_merged.head()
````

````python
# creaermos datos de ejemplo con sample
df100 = df.sample(100, random_state=1)
df102 = df.sample(100, random_state=2)

#uninermos los datos uno debajo del otro

df100.append(df102)
````

````python
# lo podemos realizar igual con concat
pd.concat([df100, df102], axis=1)
````

**Nota:**  
<font color=read> Si quisieramos unir con concat en vez de uno debajo del otro realizar join es como sigue:   
```python
pd.concat([df100,df102], axis = 1)
pd.concat([df100,df102], axis = 'columns')
    ```
