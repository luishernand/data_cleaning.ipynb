**La limpieza de datos, data cleansing o scrubbing** es un proceso necesario para asegurar la calidad de los datos que se emplearán para analytics. Este paso es fundamental para minimizar el riesgo que supondría el basar la toma de decisiones en información poco precisa, errónea o incompleta.  
El objetivo final de cualquier acción de data cleansing es mejorar la confianza de la organización
en sus datos. 

### Libreria a utilizar

````python
import pandas as pd
````

### Carga de dataset

````python
datos= pd.read_csv('vehicles.csv', nrows = 70000)
````

### Verificar y remover duplicados

````python
#verifica si tiene registros duplicados
datos.duplicated().sum()
````

Remover Duplicados

```python
datos.drop_duplicated(inplace=True)
```

### Verificar si hay registros nulos

````python
#isnull() esto nos trae la información de forma booleana
datos.isnull()
````

````python
# para verificar la cantidad
datos.isnull().sum()
````

````python
#ver el procentaje
datos.isnull().mean()
````

Otra forma de verificar los registros nulos es:

```python
nombre_archivo.isna()
```
Este comando es el mismo que isnull

````python
datos.isna().sum()
````

````python
datos.isna().mean()
````

### Remover datos nulos

````python
#lo removemos en base a un porcentaje 
thresh = len(datos)*.6

datos.dropna(thresh=thresh, axis ='columns')
````

````python
#remover rows
datos.dropna(thresh=21, axis=0)
````

### LLenar campos vacios

````python
#llenar los datos con la mediana
datos.fillna(datos.odometer.median())
````

````python
#rellenar los datos con el promedio
datos.odometer.fillna(datos.odometer.mean())
````

````python
# Rellenar los datos con cero
datos.odometer.fillna(0)
````

````python
#rellena los datos con el metodo foward
datos.fillna('ffill')
````

````python
rellena los datos con el metodo back
datos.odometer.fillna('bfill')
````

### Inferir en los datos

````python
# cambiar los datos a mayuscula
datos.region.apply(lambda x: x.upper())
````

````python
#convertirlo a minuscula
datos.region.apply(lambda x: x.lower())
````

````python
# otra forma de convertirlos a mayuscula es
datos.region.str.upper()
````

````python
datos.region.str.lower()
````

### Cambiar el tipo de datos

````python
datos['cylinders'] = pd.to_numeric(datos.cylinders, errors='coerce')
datos.dtypes
````

### Visualizar Outliers y Removerlos

````python
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
datos.boxplot('price')
````

````python
datos.boxplot('odometer')
````

````python
# Removerlos en base al quantile
df_outliers = datos[(datos.price < datos.price.quantile(.995)) & (datos.price > datos.price.quantile(.005))]
df_outliers.boxplot('price')
````

````python
df_outliers.price.hist()
````

````python
dat_outliers = datos[(datos.price< 9950) & (datos.price >0)]
dat_outliers.boxplot('price')
````
